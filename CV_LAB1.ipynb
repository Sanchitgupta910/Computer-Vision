```python
#convert img1 from BGR to RGB format
img1 = cv2.cvtColor(img1, cv2.COLOR_BGR2RGB)

#convert img1 from nparray to PIL format
img1_pil = Image.fromarray(img1)
```

```python
#convert img1 from jpg to png format
cv2.imwrite("img2.png", img1)

#convert img1 from jpg to tiff format
cv2.imwrite("img2.tiff", img1)

#convert img1 from jpg to webp format
cv2.imwrite("img2.webp", img1)

#convert img1 from jpg to bitmap format
cv2.imwrite("img2.bmp", img1)
```
