```python
#gaussian noise
import cv2
import numpy as np
from google.colab.patches import cv2_imshow
img = cv2.imread('/content/drive/MyDrive/image3.jpg')
gauss = np.random.normal(0,1,img.size)
gauss = gauss.reshape(img.shape[0],img.shape[1],img.shape[2]).astype('uint8')
# Add the Gaussian noise to the image
img_gauss = cv2.add(img,gauss)
# Display the image
cv2_imshow(img_gauss)
```

```python
from skimage.util import random_noise
img = cv2.imread("/content/drive/MyDrive/image3.jpg")
# Add salt-and-pepper noise to the image.
noise_img = random_noise(img, mode='s&p',amount=0.1)
noise_img = np.array(255*noise_img, dtype = 'uint8')

# Display the noise image
cv2_imshow(noise_img)
```

```python
def compare_image(image1, image2):
  plt.figure(figsize=(9,9))
  plt.subplot(1,2,1)
  plt.imshow(image1)
  plt.title('Orignal')
  plt.axis('off')

  plt.subplot(1,2,2)
  plt.imshow(image2)
  plt.title('Modified')
  plt.axis('off')

  plt.tight_layout()
```

```python
img = img_gauss
img2 = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
compare_image(img2,img)
```

```python
park = noise_img
park = cv2.cvtColor(park,cv2.COLOR_BGR2RGB)
edge = cv2.Canny(park,100,200)
compare_image(park,edge)
```

```python
girl = cv2.imread('/content/drive/MyDrive/image3.jpg')
girl = cv2.cvtColor(park,cv2.COLOR_BGR2RGB)
edge = cv2.Canny(girl,100,200)
compare_image(girl,edge)
```
